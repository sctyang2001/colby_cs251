> * `transform()`: Uses a homogeneous transformation matrix (passed as a parameter) to transform self.data.
5. Normalization:
> * `normalize_together()`: Uses homogeneous transformation matrices to normalize all the features of self.data together, using the global min and max.
> * `normalize_separately()`: Uses homogeneous transformation matrices to normalize each feaure separately, using its own local min and max.

### 2. Perform different matrix transformations on data

- Normalize by Z-score rather than min/max.
- "Whiten" a dataset.
- Implement normalize together and separately using numpy vectorization/broadcasting. Compare the approaches in efficiency (time and compare the two implementations).

```python
class Extension(transformation.Transformation):

    def normalizeZScoreTogether(self):
        '''Normalize z-scores with a together-ed manner.'''
        mean = np.mean(self.data.data)
        sd = np.std(self.data.data)
        trans_m = self.translation_matrix(np.zeros(self.data.get_num_dims())-mean)
        scale_m = self.scale_matrix(np.zeros(self.data.get_num_dims())+1/sd)
        C = scale_m @ trans_m
        return self.transform(C)
```

```python
zsTogether = Extension(iris) # Normalize by Z-Score together
zsTogether.project(zsTogether.data.get_headers())
zsTogether.normalizeZScoreTogether()
zsTogether.pair_plot(zsTogether.data.get_headers(),fig_sz=(18, 18), title = 'Z-Score Normalized (Together)')
plt.show()
```

```python
class Extension(transformation.Transformation):
    
    def normalizeZScoreSeparately(self):
        '''Normalize z-scores with a seperated manner.'''
        mean = self.mean(self.data.get_headers()) 
        sd = self.std(self.data.get_headers())
        trans_m = self.translation_matrix(-mean)
        scale_m = self.scale_matrix(1/sd)
        C = scale_m @ trans_m
        return self.transform(C)
```

```python
zsSeperate = Extension(iris) # Normalize by Z-Score seperately
zsSeperate.project(zsSeperate.data.get_headers())
zsSeperate.normalizeZScoreSeparately()
zsSeperate.pair_plot(zsSeperate.data.get_headers(),fig_sz=(18, 18), title = 'Z-Score Normalized (Seperately)')
plt.show()
```

### 3. Implement and use 2D rotation

```python
class Extension(transformation.Transformation):
  
    def rotationMatrix2D(self, angle):
        rad = np.deg2rad(angle)
        return np.array([[np.cos(rad), -np.sin(rad),0], [np.sin(rad), np.cos(rad), 0], [0,0,1]])

    def translationMatrix2D(self, xtrans, ytrans):
        T = np.eye(3)
        T[0,2] = xtrans
        T[1,2] = ytrans
        return T

    def rotate(self, degrees):
        center = np.mean(self.data.data, axis=0) # find the center first
        tDatah = self.get_data_homogeneous()
        R = self.rotationMatrix2D(degrees) 
        T1 = self.translationMatrix2D( -center[0], -center[1])
        T2 = self.translationMatrix2D( center[0], center[1])
        rotated = (T2@R@T1@tDatah.T).T
        self.data.data = rotated
        return rotated
```

```python
rotation2D = Extension(iris)
rotation2D.project(['sepal_length', 'petal_length'])
rotation2D.normalize_separately()
rotation2D.scatter('sepal_length', 'petal_length', title = 'Original Matrix', fig_sz=(10,10))
rotation2D.rotate(90)
rotation2D.scatter('sepal_length', 'petal_length', title = 'Matrix Rotated by 90 degrees', fig_sz=(10,10))
plt.show()
```
