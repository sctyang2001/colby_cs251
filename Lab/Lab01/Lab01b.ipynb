### 2a) Plot the data in a single plot

Using the [MatPlotLib](https://matplotlib.org/) documentation, visualize the dataset with Matplotlib. All charts should include a title, axis labels and a legend when applicable.

1. Import `lab1.csv` into a ndarray using genfromtxt.

2. Create a cartesian line graph of column 2 (index 2) on the X axis vs column 1 (index 1) on the Y axis using a red dashed line. [matplotlib.pyplot introduction](https://matplotlib.org/stable/tutorials/introductory/pyplot.html)
    1. Include a title (Lab 1 data) and axis labels (column 1 and column 2).
    2. Checkout [Controlling line properties](https://matplotlib.org/stable/tutorials/introductory/pyplot.html) for more information about setting [linestyle](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html), [marker](https://matplotlib.org/stable/api/markers_api.html) and [color](https://matplotlib.org/stable/gallery/color/named_colors.html).


```python
import matplotlib.pyplot as plt
import numpy as np

plotLab = np.genfromtxt("data/lab1.csv", delimiter=",")
```

```python
plt.figure(figsize=(5,5))
plt.plot(plotLab[:,2],plotLab[:,1], 'r*')
plt.xlabel('column 2')
plt.ylabel('column 1')
plt.title('Lab 1 data')
plt.show()
```

### 2b) Plot the data in multiple subplots

Create 6 graphs in 2 columns and 3 rows using [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html).  \
*Note: this can be done with a for loop*

1. Set figure size to 12x8. [Figure reference](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.figure.html) 
1. Row 1 is two subplots (a scatterplot and a line graph) of column 1 graphed on the x-axis VS column 0 graphed on the y-axis
2. Row 2 is two subplots (a scatterplot and a line graph) of column 2 (x-axis) VS column 1 (y-axis)
3. Row 3 is two subplots (a scatterplot and a line graph) of column 2 (x-axis) VS column 0 (y-axis)
4. Include a title (eg. column 1 vs column 2) and axis labels (eg. column 0) for each subplot.
5. To insure there is enough space between subplots use [subplots_adjust](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)


```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(plotLab[:,1],plotLab[:,0], 'r*')
plt.xlabel('column 1')
plt.ylabel('column 0')
plt.title('column 1 vs column 0')

plt.subplot(3, 2, 2)
plt.plot(plotLab[:,1],plotLab[:,0], 'y-')
plt.xlabel('column 1')
plt.ylabel('column 0')
plt.title('column 1 vs column 0')

plt.subplot(3, 2, 3)
plt.plot(plotLab[:,2],plotLab[:,1], 'b*')
plt.xlabel('column 2')
plt.ylabel('column 1')
plt.title('column 2 vs column 1')

plt.subplot(3, 2, 4)
plt.plot(plotLab[:,2],plotLab[:,1], 'p-')
plt.xlabel('column 2')
plt.ylabel('column 1')
plt.title('column 2 vs column 1')

plt.subplot(3, 2, 5)
plt.plot(plotLab[:,2],plotLab[:,0], 'r*')
plt.xlabel('column 2')
plt.ylabel('column 0')
plt.title('column 2 vs column 0')

plt.subplot(3, 2, 6)
plt.plot(plotLab[:,2],plotLab[:,0], 'g-')
plt.xlabel('column 2')
plt.ylabel('column 0')
plt.title('column 2 vs column 0')

plt.subplots_adjust(hspace = .5)

plt.show()
```
